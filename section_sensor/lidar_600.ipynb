## 2.4 複雑な分布


### 2.4.1 条件付き確率


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(
    "sensor_data_600.txt",
    delimiter=" ",
    header=None,
    names=("date", "time", "ir", "lidar"),
)

data["lidar"].hist(bins=max(data["lidar"]) - min(data["lidar"]), align="left")
plt.show()
```

```python
data["lidar"].plot()
plt.show()
```

```python
data["hour"] = [e // 10000 for e in data.time]
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(11).hist()
plt.show()
```

### 2.4.2 同時確率と加法定理、乗法定理


```python
each_hour = {i: d.lidar.get_group(i).value_counts().sort_index() for i in range(24)}
freqs = pd.concat(each_hour, axis=1)
freqs = freqs.fillna(0).sort_index()
probs = freqs / len(data)

probs
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
sns.jointplot(x=data["hour"], y=data["lidar"], kind="kde")
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum())
p_t.plot()
p_t.transpose()
```

```python
p_t.sum()
```

```python
p_z = pd.DataFrame(probs.transpose().sum())
p_z.plot()
p_z.transpose()
```

```python
p_z.sum()
```

```python
cond_z_t = probs/p_t[0]
cond_z_t
```

```python
(cond_z_t[6]).plot.bar(color='blue', alpha=0.5)
(cond_z_t[14]).plot.bar(color='orange', alpha=0.5)
plt.show()
```
